# 5.0 Preparação de Dados

## Separação entre treino e teste

```python
X = df4.drop('response', axis = 1 )
y = df4['response'].copy() 

x_train, x_validation, y_train, y_validation = ms.train_test_split( X, y, test_size=0.20 )

#20% dos dados serão usados para validação. 
#Esses conjuntos são divididos de forma aleatória, preservando a distribuição dos dados originais 
#entre as classes, caso existam.

df5 = pd.concat([x_train, y_train], axis = 1) 

#como as series y_train e y_validation já tinham nome de response, que vem do nome da coluna, 
#ao fazer o concat, automaticamente esse nome vai pro nome da coluna do dataframe resultante. 
```

## Standardization

```python
#annual_premium

from sklearn import preprocessing as pp 
ss = pp.StandardScaler()

df5['annual_premium'] = ss.fit_transform(df5[['annual_premium']].values)       
pickle.dump(ss, open('C:\\Users\\Matheus\\repos\\health_insurance_cross_sell_oficial\\src\\features\\annual_premium_scaler.pkl', 'wb' ))
```

## Rescaling

```python
#salvando encoder dentro de uma variável para depois utilizar no conjunto de teste. 
mms_age = pp.MinMaxScaler()
mms_vintage = pp.MinMaxScaler() 

#não se aproximam de uma normal 
#Age 

df5['age'] = mms_age.fit_transform(df5[['age']].values) 
pickle.dump(mms_age, open('C:\\Users\\Matheus\\repos\\health_insurance_cross_sell_oficial\\src\\features\\age_scaler.pkl', 'wb'))

# vintage
df5['vintage'] = mms_vintage.fit_transform(df5[['vintage']].values) 
pickle.dump(mms_vintage, open('C:\\Users\\Matheus\\repos\\health_insurance_cross_sell_oficial\\src\\features\\vintage_scaler.pkl', 'wb'))
```

```python
df5
```

## Encoder

```python
#gender - #Pode utilizar: One Hot Enconding / Frequency Encodging / Target Encoding / Weighted Target Encoding 
#foi utilizado target enconding 
target_encode_gender = df5.groupby('gender')['response'].mean() 
df5.loc[:, 'gender'] = df5['gender'].map(target_encode_gender) #substitui male por 0.138411 e female por 0.103902
pickle.dump(target_encode_gender, open('C:\\Users\\Matheus\\repos\\health_insurance_cross_sell_oficial\\src\\features\\target_encode_gender_scaler.pkl', 'wb'))

#Se é maior que 8 valores únicos, esquece one hot enconding. Para não aumentar a dimensionalidade da tua database a ponto de 
#ferrar teu modelo. 
#Ir testando de ciclo em ciclo caso o resultado saia muito ruim. 

#region_code = Target Encoding/Frequency Encoding 
target_encode_region_code = df5.groupby('region_code')['response'].mean() #substitui pela média das respostas de cada região. 
df5.loc[:, 'region_code'] = df5['region_code'].map(target_encode_region_code)
pickle.dump(target_encode_region_code, open('C:\\Users\\Matheus\\repos\\health_insurance_cross_sell_oficial\\src\\features\\target_encode_region_code_scaler.pkl', 'wb'))

#vehicle_age - Order Enconding funciona bem pois eu tenho uma ordem. / One Hot Encoding / Frequency 
#foi usado one hot encoding 
df5 = pd.get_dummies(df5, prefix='vehicle_age', columns=['vehicle_age'])

#policy_sales_channel = Target Encoding / Frequency Encoding 
#substitui pela frequência de cada canal no dataframe 
fe_policy_sales_channel = df5.groupby('policy_sales_channel').size() / len(df5) #len retorna o número de linhas do dataframe
df5.loc[:, 'policy_sales_channel'] = df5['policy_sales_channel'].map(fe_policy_sales_channel) 
pickle.dump(fe_policy_sales_channel, open('C:\\Users\\Matheus\\repos\\health_insurance_cross_sell_oficial\\src\\features\\fe_policy_sales_channel_scaler.pkl', 'wb'))
#quando tu usa map aqui você diz: quando for valor tal (coluna 0) , você substitui por valor tal (coluna 1).
```

## 5.4 Preparação de Validação

```python
# gender
x_validation.loc[:, 'gender'] =  x_validation.loc[:, 'gender'].map( target_encode_gender )

# age
x_validation.loc[:, 'age'] = mms_age.fit_transform( x_validation[['age']].values )

# region_code
x_validation.loc[:, 'region_code'] = x_validation.loc[:, 'region_code'].map( target_encode_region_code )

# vehicle_age
x_validation = pd.get_dummies( x_validation, prefix='vehicle_age', columns=['vehicle_age'] )

# annual_premium
x_validation.loc[:, 'annual_premium'] = ss.transform( x_validation[['annual_premium']].values )

# policy_sales_channel
x_validation.loc[:, 'policy_sales_channel'] = x_validation['policy_sales_channel'].map( fe_policy_sales_channel )

# vintage
x_validation.loc[:, 'vintage'] = mms_vintage.transform( x_validation[['vintage']].values )

# fillna
x_validation = x_validation.fillna( 0 )
```

# 8.0. Model Performance

```python
def precision_at_k( data, k=2000 ):
    # reset index
    data = data.reset_index( drop=True )

    # create ranking order
    data['ranking'] = data.index + 1 

    data['precision_at_k'] = data['response'].cumsum() / data['ranking']

    return data.loc[k, 'precision_at_k']
```

```python
def recall_at_k( data, k=2000 ):
    # reset index
    data = data.reset_index( drop=True )

    # create ranking order
    data['ranking'] = data.index + 1 

    data['recall_at_k'] = data['response'].cumsum() / data['response'].sum()

    return data.loc[k, 'recall_at_k']
```

## Deploy to production

```python
import pandas as pd
import pickle
import inflection

class HealthInsurance(): 
    #transformando nomes de coluna para snake case    
    def __init__(self): 
        self.annual_premium_scaler = pickle.load(open('src/features/annual_premium_scaler.pkl', 'rb'))
        self.age_scaler = pickle.load(open('src/features/age_scaler.pkl', 'rb'))
        self.vintage_scaler = pickle.load(open('src/features/vintage_scaler.pkl', 'rb'))
        self.target_encode_gender_scaler = pickle.load(open('src/features/target_encode_gender_scaler.pkl', 'rb'))
        self.target_encode_region_code = pickle.load(open('src/features/target_encode_region_code_scaler.pkl', 'rb'))
        self.fe_policy_sales_channel = pickle.load(open('src/features/fe_policy_sales_channel_scaler.pkl', 'rb'))
        
    def data_cleaning(self, df1):  
        
        cols_old = df1.columns

        snakecase = lambda x: inflection.underscore(x)

        df1.columns = list(map(snakecase, cols_old))
        
        return df1 

    
    def feature_engineering(self, df2): 
        
        df2['vehicle_age'] = df2['vehicle_age'].apply(lambda x: 'over_2_years' if x == '> 2 Years' 
                                                      else 'between_1_and_2_years' if x == '1-2 Year' 
                                                      else 'bellow_1_year')

        df2['vehicle_damage'] = df2['vehicle_damage'].apply(lambda x: 1 if x == 'Yes' 
                                                      else 0 )

        return df2 

    def data_preparation(self, df4):  

        #annual_premium

        df4['annual_premium'] = self.annual_premium_scaler.fit_transform(df4[['annual_premium']].values)

        #não se aproximam de uma normal 
        
        #Age

        df4['age'] = self.age_scaler.fit_transform(df4[['age']].values) 

        # vintage

        df4['vintage'] = self.age_scaler.fit_transform(df4[['vintage']].values) 
        
        #gender - Target Enconding 
        df4.loc[:, 'gender'] = df4['gender'].map(self.target_encode_gender_scaler) #substitui male por 0.138411 e female por 0.103902

        #region_code - Target Encoding
        df4.loc[:, 'region_code'] = df4['region_code'].map(self.target_encode_region_code)

        #vehicle_age - One Hot Encoding 
        df4 = pd.get_dummies(df4, prefix='vehicle_age', columns=['vehicle_age'])

        #policy_sales_channel - Frequency Encoding 
        df4.loc[:, 'policy_sales_channel'] = df4['policy_sales_channel'].map(self.fe_policy_sales_channel)    
        
        # 6.0 Feature Selection 
        
        cols_selected = ['annual_premium', 'vintage', 'age', 'region_code', 'vehicle_damage', 'previously_insured', 'policy_sales_channel']
        
        return df4[cols_selected]
    
    def get_prediction(self, model, original_data, test_data): 
        #model prediction 
        pred = model.predict_proba(test_data) 

        #join prediciton into original data 
        original_data['prediction'] = pred 

        return original_data.to_json(orient = 'records', date_format = 'iso')
```

```python
import os 
import pickle
import pandas as pd
from flask             import Flask, request, Response
from health_insurance.HealthInsurance import HealthInsurance
import json

# loading model
model = pickle.load(open('models/model_lgbm2.pkl', 'rb'))

# initialize API
app = Flask( __name__ )

@app.route( '/predict', methods=['POST'] )
def health_insurance_predict():
    test_json = request.get_json()
   
    if test_json: # there is data
        if isinstance( test_json, dict ): # unique example
            test_raw = pd.DataFrame( test_json, index=[0] )
            
        else: # multiple example
            test_raw = pd.DataFrame( test_json, columns=test_json[0].keys() )
            
        # Instantiate HealthInsurance(): 

        pipeline = HealthInsurance()
        
        # data cleaning
        df1 = pipeline.data_cleaning( test_raw )
        
        # feature engineering
        df2 = pipeline.feature_engineering( df1 )
        
        # data preparation
        df3 = pipeline.data_preparation( df2 )
        
        # prediction
        df_response = pipeline.get_prediction( model, test_raw, df3 )
        
        return df_response       
        
    else:
        return Response( '{}', status=200, mimetype='application/json' )

#teste local
if __name__ == '__main__':
    app.run(host='0.0.0.0', port = 5000)
    
#deploy
#if __name__ == '__main__':
    #port = os.environ.get('PORT', 5000)
    #app.run(host='0.0.0.0', port = port)
```
